```python
import pandas as pd
```

```python
df=pd.read_csv('homeprices.csv')
```

```python
df
```

```python
dummies=pd.get_dummies(df.town)
```

```python
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
```

```python
merged
```

```python
merged.drop('town',axis='columns',inplace=True)
```

```python
merged
```

```python
merged.drop('west windsor',axis='columns',inplace=True)
```

```python
merged
```

```python
merged.area=df.area/(df.area.max())
```

```python
x=merged.drop('price',axis='columns')
```

```python
y=df.price
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```
